```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import random

# 读取 h5 文件
h5_path = '/root/autodl-tmp/EventMamba-main/data/DVS-Lip/test.h5'
with h5py.File(h5_path, 'r') as f:
    data = f['data'][:]   # (样本数, 4096, 3)
    labels = f['label'][:]
print(f'点云样本数: {data.shape[0]}, 每个点云形状: {data.shape[1:]}')

# 随机抓取一个点云
idx = random.randint(0, data.shape[0] - 1)
point_cloud = data[idx]
label = labels[idx]
print(f'随机选中样本编号: {idx}, 标签: {label}')

# 可视化点云
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(point_cloud[:, 0], point_cloud[:, 1], point_cloud[:, 2], s=1)
ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_zlabel('y')
ax.set_title(f'DVS-Lip Point Cloud (label={label})')
plt.show()
```
